## K Means Clustering

#### K Means Clustering without PreProcessing

```python
kmeans_without_preprocessing = setup(data = dataset_new)
```

```python
print("For Cluster = 3")
_3clustermodel = create_model('kmeans', num_clusters = 3)
_3clusterprediction = assign_model(_3clustermodel)

print("For Cluster = 4")
_4clustermodel = create_model('kmeans', num_clusters = 4)
_4clusterprediction = assign_model(_4clustermodel)

print("For Cluster = 5")
_5clustermodel = create_model('kmeans', num_clusters = 5)
_5clusterprediction = assign_model(_5clustermodel)
```

```python
print("For C = 3")
plot_model(_3clustermodel, plot='cluster')
plot_model(_3clustermodel, plot='silhouette')
plot_model(_3clustermodel, plot='distribution')
print("For C = 4")
plot_model(_4clustermodel, plot='cluster')
plot_model(_4clustermodel, plot='silhouette')
plot_model(_4clustermodel, plot='distribution')
print("For C = 5")
plot_model(_5clustermodel, plot='cluster')
plot_model(_5clustermodel, plot='silhouette')
plot_model(_5clustermodel, plot='distribution')
```

To Find the optimal number of Clusters 

```python
plot_model(_5clustermodel, plot='elbow')
```
